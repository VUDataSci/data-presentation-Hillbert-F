```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
current_directory = os.getcwd()
print(f"Current working directory: {current_directory}")
```

```python
xls_path1 = current_directory + './data/raw/API_FP.CPI.TOTL.ZG_DS2_en_excel_v2_5994828.xls'
xls_path2 = current_directory + './data/raw/API_NY.GNP.ATLS.CD_DS2_en_excel_v2_5995505.xls'
```

```python
df1 = pd.read_excel(xls_path1, 'Data')
df2 = pd.read_excel(xls_path1, 'Metadata - Countries')
df3 = pd.read_excel(xls_path2, 'Data')
df4 = pd.read_excel(xls_path2, 'Metadata - Countries')
```

```python
# Merging the two dataframes based on 'Country Code' ensuring the information matches correctly
merged_df = pd.merge(df1, df2[['Country Code', 'IncomeGroup', 'Region']], on='Country Code', how='left')

# Reordering the columns to place 'IncomeGroup' and 'Region' after 'Country Code'
cols = merged_df.columns.tolist()
incomegroup_index = cols.index('IncomeGroup')
region_index = cols.index('Region')
cols = cols[:3] + [cols[incomegroup_index]] + [cols[region_index]] + cols[3:incomegroup_index] + cols[incomegroup_index+1:region_index] + cols[region_index+1:]
merged_df = merged_df[cols]

merged_df.head()
```

```python
# Moving rows where 'Region' is NaN to the bottom
merged_df = merged_df.sort_values(by='Region', na_position='last')

merged_df.head()
```

```python
# Merging the two dataframes based on 'Country Code' ensuring the information matches correctly
merged_df_2 = pd.merge(df3, df4[['Country Code', 'IncomeGroup', 'Region']], on='Country Code', how='left')

# Reordering the columns to place 'IncomeGroup' and 'Region' after 'Country Code'
cols_2 = merged_df_2.columns.tolist()
incomegroup_index_2 = cols_2.index('IncomeGroup')
region_index_2 = cols_2.index('Region')
cols_2 = cols_2[:3] + [cols_2[incomegroup_index_2]] + [cols_2[region_index_2]] + cols_2[3:incomegroup_index_2] + cols_2[incomegroup_index_2+1:region_index_2] + cols_2[region_index_2+1:]
merged_df_2 = merged_df_2[cols_2]

merged_df_2.head()
```

```python
# Moving rows where 'Region' is NaN to the bottom
merged_df_2 = merged_df_2.sort_values(by='Region', na_position='last')

merged_df_2.head()
```

```python
xlsx_path_new1 = str(current_directory) + './data/cleaned/Country_CPI.xlsx'
xlsx_path_new2 = str(current_directory) + './data/cleaned/Area_CPI.xlsx'
xlsx_path_new3 = str(current_directory) + './data/cleaned/Gross_National_Income.xlsx'
```

```python
# Splitting the merged dataframe into two dataframes based on the presence of NaN in the 'Region' column
df_countries = merged_df[merged_df['Region'].notna()]
df_areas = merged_df[merged_df['Region'].isna()]

# Renaming the 'Country Name' column to 'Area Name' for the areas dataframe
df_areas = df_areas.rename(columns={"Country Name": "Area Name"})

df_countries.to_excel(xlsx_path_new1)
df_areas.to_excel(xlsx_path_new2)
merged_df_2.to_excel(xlsx_path_new3)
```
